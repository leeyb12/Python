```python
import requests as req

# 1) ê¸°ë³¸ API í˜¸ì¶œ â€” ì „ì²´ ë¶€ì‘ìš© ë°˜ì‘ ê±´ìˆ˜
base_url = "https://api.fda.gov/drug/event.json?count=patient.reaction.reactionmeddrapt.exact"
data = req.get(base_url).json()

results = data["results"]

# í•„ìš”í•œ ë¶€ì‘ìš©ì„ ì°¾ëŠ” helper í•¨ìˆ˜
def find_count(term):
    for item in results:
        if item["term"] == term:
            return item["count"]
    return 0

# ğŸŸ¡ NAUSEA ë³´ê³  ê±´ìˆ˜
nausea_count = find_count("NAUSEA")

# ğŸŸ¡ VOMITING(êµ¬í† ) ë³´ê³  ê±´ìˆ˜
vomit_count = find_count("VOMITING")

# ğŸŸ¡ COVID-19 ê´€ë ¨ ë³´ê³  ê±´ìˆ˜
covid_count = find_count("COVID-19") + find_count("COVID 19")

# ğŸŸ¡ MYOCARDIAL INFARCTION(ì‹¬ê·¼ê²½ìƒ‰)
mi_count = find_count("MYOCARDIAL INFARCTION")

# 2) ë¹„êµ íŒë‹¨
comparison = "ì½”ë¡œë‚˜ê°€ ë” ë§ìŒ" if covid_count > mi_count else "ì‹¬ê·¼ê²½ìƒ‰ì´ ë” ë§ìŒ"

# 3) NAUSEA vs VOMITING ë¹„êµ
if nausea_count > vomit_count:
    nv_msg = f"NAUSEAê°€ ë” ë§ìŒ (ì°¨ì´: {nausea_count - vomit_count} ê±´)"
elif vomit_count > nausea_count:
    nv_msg = f"VOMITINGì´ ë” ë§ìŒ (ì°¨ì´: {vomit_count - nausea_count} ê±´)"
else:
    nv_msg = "NAUSEAì™€ VOMITINGì€ ë™ì¼í•œ ê±´ìˆ˜"

# 4) meta ì •ë³´ì˜ disclaimerì— "FDA" í¬í•¨ ì—¬ë¶€ í™•ì¸
disclaimer = data.get("meta", {}).get("disclaimer", "")
contains_fda = "FDA" in disclaimer

# ì¶œë ¥
print("ğŸ“Š NAUSEA ë³´ê³  ê±´ìˆ˜   :", nausea_count)
print("ğŸ“Š COVID-19 ë³´ê³  ê±´ìˆ˜:", covid_count)
print("ğŸ“Š ì‹¬ê·¼ê²½ìƒ‰ ë³´ê³  ê±´ìˆ˜ :", mi_count)
print("ğŸ” ë¹„êµ íŒë‹¨         :", comparison)
print("ğŸ“Š NAUSEA vs VOMITING :", nv_msg)
print("ğŸ“ Disclaimer ì•ˆì— 'FDA' í¬í•¨ ì—¬ë¶€:", contains_fda)
```
