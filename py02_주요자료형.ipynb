```python
# reverse - 리스트 뒤집기
a = [5, 43, 5, 6, 56, 3, 2, 4, 1]
b = ['gf', 'f', 'u', 'a', 'e', 'f', 'z', 'c']
print(a[::-1])
print(b[::-1])
a.reverse()
b.reverse()
print(a)
print(b)
```

```python
# sorted - 원본 유지하기
aa = [66, 7, 8, 99, 0]
print(sorted(aa))
print(sorted(aa, reverse=True))
print(aa)
```

```python
# 이전에 알아낸 실시간 비트코인 BTC / USD 가격으로 
# 실시간 환율을 계산하여 
# 지금 비트코인 가격을 원화로 표시하시오.
```

```python
import requests
import re

# 1. BTC / USD 실시간 가격 (Binance)
btc_url = "https://api4.binance.com/api/v3/ticker/price?symbol=BTCUSDT"
btc_data = requests.get(btc_url).json()

# 소수점 앞부분만 사용 (문자열 처리)
btc_usd = int(btc_data["price"].split(".")[0])

# 2. USD / KRW 실시간 환율 (한국수출입은행)
fx_url = "https://oapi.koreaexim.go.kr/site/program/financial/exchangeJSON?authkey=2ePkmyjmWEOy1ym1qC6Dlba1dcJ8J2m9&data=AP01"
fx_text = requests.get(fx_url).text   # 🔥 여기만 수정

# USD 환율 추출
usd_to_krw = int(
    re.search(
        r'"cur_unit":"USD".*?"deal_bas_r":"([\d,]+)',
        fx_text
    ).group(1).replace(",", "")
)

# 3. 비트코인 원화 가격 계산
btc_krw = btc_usd * usd_to_krw
won = str(btc_krw)

print("비트코인 가격 기반 실시간 환율 계산")
print(f"BTC / USD : {btc_usd:,} 달러")
print(f"USD / KRW : {usd_to_krw:,} 원")
print("비트코인 원화 가격:", won[:-8] + "억 " + won[-8:-4] + "만 " + won[-4:] + "원")
```

```python
# access_logs = ["192.168.0.1", "192.168.0.2", "192.168.0.1", "10.0.0.1", "192.168.0.2", "192.168.0.5"]
# '''
# 요구 사항:
# access_logs를 set을 사용하여 중복을 제거하십시오.
# 새로운 관리자 IP "192.168.0.99"를 추가하십시오.
# 최종 고유 IP 개수를 출력하십시오.
# '''
```

```python

access_logs = ["192.168.0.1", "192.168.0.2", "192.168.0.1", "10.0.0.1", "192.168.0.2", "192.168.0.5"]

# set을 사용하여 중복 제거
unique_ips = set(access_logs)
print(unique_ips)
# 새로운 관리자 IP 추가
unique_ips.add("192.168.0.99")
print(unique_ips)
# 최종 고유 IP 개수 출력
print("최종 고유 IP 개수:", len(unique_ips))
```

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
# 교집합 & 
print("교집합", s1 & s2) # s1.intersection(s2) 가능
# 합집합 |
print("합집합", s1 | s2) # s1.union(s2) 가능
# 차집합 -
print("차집합", s1 - s2) # s1.difference(s2) 가능
```

### 집합 자료형 관련 함수

```python
# add 집합자료형 값 1개 추가
s1 = set([1, 2, 3])
print(s1)
s1.add(4)
print(s1)
```

```python
# update 집합자료형 값 여러개 추가하기
s1 = set([1, 2, 3])
print(s1)
s1.update([2, 3, 4, 5, 6, 7])
print(s1)
```

```python
# remove 집합자료형 특정값 제거하기
s1 = set([1, 2, 3])
print(s1)
s1.remove(2) # index가 아닌 값으로 선택 
print(s1)
```

**존재하지 않는 값을 찾을 때 오류**



```python
# remove 집합자료형 특정 값 제거하기 
s1 = set([1, 2, 3])
print(s1)
s1.discard(0) # 없는 값이어도 오류가 발생하지 않음
print(s1)
```

```python
# clear 집합자료형 내 모든 값 제거
s1 = set([1, 2, 3])
s1.clear()
print(s1)
```

```python
# or 연산자
print("True or True: ->", True or True)
print("True or False: ->", True or False)
print("False or True: ->", False or True)
print("False or False: ->", False or False)
```
