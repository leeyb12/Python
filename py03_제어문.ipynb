```python
# if, elif ~~
pocket = ['paper', 'key']
card = False
if 'money' in pocket:
    print("택시 타라")   
elif card == True:
    print("택시 신나게 타라")
else:
    print("걸어가라")
```

```python
# if만 사용
pocket = ['paper', 'key']
card = True
if 'money' in pocket:
    print("택시 빨리 타라")   
if card == True:
    print("택시 신나게 타라")
if not ('money' in pocket) and not card == True:
    print("걸어가라")
```

## while 문 **

```python
check = 0
while check < 10 :
    check = check + 1 # check += 1
    print(f"{check}번째 가즈아~~!!")
    if check == 10:
        print("구조대 탈출!")
```

```python
treeHit = 0
while treeHit < 10:
    treeHit = treeHit + 1
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다.")
```

```python
# 응용 (1 ~ 6 주사위)
import random as ran

n = 0
box = []
while n < 100:
    num = ran.randint(1, 6)
    n += 1
    box.append(num)

# 각 눈금별 등장 횟수
c1 = box.count(1)
c2 = box.count(2)
c3 = box.count(3)
c4 = box.count(4)
c5 = box.count(5)
c6 = box.count(6)

print(f"주사위 눈금 1번이 {c1}번 등장하였습니다. ({c1/100*100:0.2f}%)")
print(f"주사위 눈금 2번이 {c2}번 등장하였습니다. ({c2/100*100:0.2f}%)")
print(f"주사위 눈금 3번이 {c3}번 등장하였습니다. ({c3/100*100:0.2f}%)")
print(f"주사위 눈금 4번이 {c4}번 등장하였습니다. ({c4/100*100:0.2f}%)")
print(f"주사위 눈금 5번이 {c5}번 등장하였습니다. ({c5/100*100:0.2f}%)")
print(f"주사위 눈금 6번이 {c6}번 등장하였습니다. ({c6/100*100:0.2f}%)")
```

```python
import random as ran
n = 0
box = []
cc = int(input("몇회 주사위를 던질까요?") or 100)
while n < cc:
    num = ran.randint(1, 6)
    n += 1
    box.append(num)
    # print(f"{n}번째:", num)

if cc == len(box):
    print(f"{cc}회 정상 진행 되었습니다.")
    print(f"주사위 눈금 1이 {box.count(1)}번 { ((box.count(1)/cc) * 100):0.2f}% 출현하였습니다.")
    print(f"주사위 눈금 2이 {box.count(2)}번 { ((box.count(2)/cc) * 100):0.2f}% 출현하였습니다.")
    print(f"주사위 눈금 3이 {box.count(3)}번 { ((box.count(3)/cc) * 100):0.2f}% 출현하였습니다.")
    print(f"주사위 눈금 4이 {box.count(4)}번 { ((box.count(4)/cc) * 100):0.2f}% 출현하였습니다.")
    print(f"주사위 눈금 5이 {box.count(5)}번 { ((box.count(5)/cc) * 100):0.2f}% 출현하였습니다.")
    print(f"주사위 눈금 6이 {box.count(6)}번 { ((box.count(6)/cc) * 100):0.2f}% 출현하였습니다.")
else:
    print(f"검증 결과가 이상이 있습니다. {cc}회 목표로 동작하였으나, {len(box)}으로 측정되었습니다.")
```

```python
# 로또 번호
# 1 ~ 45임의의 수 중에서 중복 없는 6개의 수를 뽑아 나열하여 분석하시오.

import random as ran 
lotto = [] # 로또 번호 저장할 리스트 
n = 0 # 뽑은 횟수 카운트 

# 중복 없는 6개 번호 뽑기 
while n < 6: 
    num = ran.randint(1, 45) 
    if num not in lotto: # 중복 체크 
        lotto.append(num) 
        n += 1 
        
print("이번 로또 번호:", lotto)
```
